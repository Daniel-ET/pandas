## **Exercise 41**

The following objects are given

* s1 = pd.Series(np.random.rand(20))
* s2 = pd.Series(np.random.rand(20))

Combine these two Series into one DataFrame and assign to df variable. Set the column names 'col1' and 'col2' respectively

```python
s1 = pd.Series(np.random.rand(20))
s2 = pd.Series(np.random.randn(20))
```

```python
df = pd.concat([s1, s2], axis=1)
df.columns = ['col1', 'col2']
df
```

### **Exercise 42**

```python
df[df['col2'].between(0.0, 1.0)]
```

```python
df[(df['col2'] >= 0.0) & (df['col2'] <= 1.0)]
```

### **Exercise 43**

Assign a new column named 'col3', which will store 1 when the value in the column col2 is non-negative and -1 opposite.

```python
df['col3'] = df['col2'].map(lambda x: 1 if x >= 0 else -1)
df
```

### **Exercise 44**

Assign a new column 'col4', which will truncate the values from the column col2 to the range [âˆ’1.0, 1.0]. In other words, for values less than -1.0 set -1.0 and for values greater than 1.0 set 1.0.

```python
df['col4'] = df['col2'].clip(-1.0, 1.0)
df
```
