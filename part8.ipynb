### **Exercise 71**

Check the distribution of the transmission column and perform the following mapping:

* url - https://storage.googleapis.com/esmartdata-courses-files/dash-course/data.csv
* 'Manual' -> 0
* 'Automatic' -> 1

print the first five rows of the transmission column

```python
url = 'https://storage.googleapis.com/esmartdata-courses-files/dash-course/data.csv'
df = pd.read_csv(url, index_col=0)
df.columns = map(str.lower, df.columns)
df.head(5)
```

### **Exercise 78**

Group this DataFrame by month and calculate the average value for the hum column. Assign the result to humidity_by_month variable and reset the index.

In response, print humidity_by_month DataFrame to the console as shown below.

```python
humidity_by_month = df.groupby('month')['hum'].mean().reset_index()
humidity_by_month
```
