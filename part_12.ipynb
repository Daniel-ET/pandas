```python
import numpy as np
import pandas as pd
import pandas_datareader as web
```

```python
amzn = web.DataReader('AMZN', data_source='stooq', start='2020-01-01', end='2022-01-01')
tsla = web.DataReader('TSLA', data_source='stooq', start='2020-01-01', end='2022-01-01')
amzn.columns = ['amzn_' + col.lower() for col in amzn.columns]
tsla.columns = ['tsla_' + col.lower() for col in tsla.columns]
```

### **Exercise 111**

Combine the amazon and tesla dataframes into one dataframe and assign to the result variable

```python
amzn.head()
```

```python
tsla.head()
```

```python
result = pd.concat([amzn, tsla], axis=1)
result.head()
```

### **Exercise 112**

Extract the following columns from the result DataFrame:

* amzn_open
* amzn_close
* tsla_open
* tsla_close

and copy to a new object called quotations.

```python
cols = ['amzn_open', 'amzn_close', 'tsla_open', 'tsla_close']
quotes = result[cols].copy()
quotes.head()
```

### **Exercise 113**

Assign two following columns to the quotations DataFrame:

* amzn_change
* tsla_change

which will be the daily percentage change of the price.

```python
quotes['amzn_change'] = (quotes['amzn_close'] / quotes['amzn_open'] - 1)
quotes['amzn_change'] = quotes['amzn_change'] * 100
quotes['tsla_change'] = (quotes['tsla_close'] / quotes['tsla_open'] - 1)
quotes['tsla_change'] = quotes['tsla_change'] * 100
quotes
```

### **Exercise 114**

Create a new column 'similarity_flag' that stores 1 when the daily percentage change for Amazon and Tesla is of the same sign, otherwise 0 as shown below.

```python
quotes['similarity_flag'] = (quotes['amzn_change'] * quotes['tsla_change'] > 0) * 1
quotes.head()
```

### **Exercise 115**

Calculate the percentage of trading sessions for which Amazon and Tesla had the same direction (similarity_flag == 1) 

```python
result = quotes['similarity_flag'].sum() / len(quotes)
print(f'{result * 100:.2f}%')
```

```python
products_dict = {
    'product_id': ['001', '004', '007', '010'],
    'name': ['mobile phone', 'laptop', 'mouse', 'tablet'],
    'price': [1490.0, 3400.0, 59.99, 999.00]
}
 
orders_day_1_dict = {
    'order_id': ['1001', '1002', '1003', '1004'],
    'product_id': ['004', '001', '001', '007'],
    'quantity': [2, 1, 1, 3]
}       
 
orders_day_2_dict = {
    'order_id': ['1005', '1006', '1007'],
    'product_id': ['010', '001', '007'],
    'quantity': [2, 1, 1]
}
```

```python
products = pd.DataFrame(products_dict)
orders_day_1 = pd.DataFrame(orders_day_1_dict)
orders_day_2 = pd.DataFrame(orders_day_2_dict)
```
