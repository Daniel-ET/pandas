```python
import pandas as pd
import numpy as np
```

```python
employees = pd.DataFrame({
    'emp_id': ['1001', '1002', '1003', '1004'],
    'first_name': ['John', 'Mark', 'Bob', 'Alice'],
    'last_name': ['Smith', 'Cook', 'Cat', 'Lee'],
    'dept_id': ['002', '001', '001', '005']
})

departments = pd.DataFrame({
    'dept_id': ['001', '002', '003', '004'],
    'dept_name': ['IT', 'Sales', 'Marketing', 'Support']
})
```

### **Exercise 121**

make an INNNER join of employees and departments DataFrames based on the dept_id column.

```python
employees.head()
```

```python
departments.head()
```

```python
result = pd.merge(employees, departments, how='inner', on='dept_id')
result
```

### **Exercise 122**

Make a LEFT join of employees and departments DataFrames based on the dept_id column. Note that the departments DataFrame doesn't have a department with dept_id == '005'.

```python
result = pd.merge(employees, departments, how='left', on='dept_id')
result
```

### **Exercise 123**

Make a RIGHT join of employees and departments DataFrames based on the dept_id column. Note that the employees object doesn't have employees for departments with dept_id of '003' and '004'.

```python
employees.head()
```

```python
departments.head()
```

### **Exercise 124**

Make an OUTER join of employees and departments DateFrames based on the dept_id column. Note that the employees DataFrmae doesn't have employees for departments with dept_id of '003' and '004'. Also note that the departments DataFrame doesn't have a department with dept_id '005'.

### **Exercise 127**

Convert the following url

* https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

into a dataframe. Then use pivot table to prepare another dataframe showing the avergae tip and total bill for each day.

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
tips.head()
```

```python
result = pd.pivot_table(tips, index='day', values=['tip', 'total_bill'], aggfunc=np.mean)
result
```

### **Exercise 128**

Create the same pivot table as above but this time index by 'day' and 'sex'

```python
result = pd.pivot_table(tips, index=['day', 'sex'], values=['tip', 'total_bill'], aggfunc=np.mean)
result
```

### **Exercise 129**

Create a pivot table showing the maximum value of the tip and total_bill columns broken down into days.

```python
result = pd.pivot_table(tips, index='day', values=['tip', 'total_bill'], aggfunc=np.max)
result
```
