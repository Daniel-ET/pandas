### **Exercise 88**

Convert the following columns:

* sex
* smoker
* region

to *category* type. In response, show some information about the dataframe to the console.

```python
cat_vars = ['sex', 'smoker', 'region']
 
for col in cat_vars:
    df[col] = df[col].astype('category')

df.info()
```
