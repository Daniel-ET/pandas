### **Exercise 21**

Load google stock price information into a pandas dataframe

```python
# loading google stock price info
df = web.DataReader('GOOG', data_source='stooq', start='2020-01-01', end='2022-01-01')
df
```

```python
#plotting close price
df['Close'].plot(figsize=(15,8))
```

### **Exercise 25**
